```python
def extract_lat_long_from_dynamic_url(short_url):
    try:
        response = requests.get(short_url)
        final_url = response.url
        print(final_url)
        # Regular expressions to find latitude and longitude in different formats
        pattern1 = r'@(-?\d+\.\d+),(-?\d+\.\d+)'
        pattern2 = r'%40(-?\d+\.\d+),(-?\d+\.\d+)'
        pattern3 = r'!3d(-?\d+\.\d+)!4d(-?\d+\.\d+)'
        # Search for the patterns in the URL
        match1 = re.search(pattern1, final_url)
        match2 = re.search(pattern2, final_url)
        match3 = re.search(pattern3, final_url)
        if match1:
            latitude = match1.group(1)
            longitude = match1.group(2)
            return latitude, longitude
        elif match2:
            latitude = match2.group(1)
            longitude = match2.group(2)
            return latitude, longitude
        elif match3:
            latitude = match3.group(1)
            longitude = match3.group(2)
            return latitude, longitude
        else:
            return None, None
    except requests.RequestException as e:
        print(f"Failed to fetch URL: {short_url}, error: {e}")
        return None, None
```

```python
def map_to_veg_nonveg(category):
    non_veg_indicators = ['Non-veg', "['Non-veg', 'Veg']", "['Non-veg']", "['Veg', 'Non-veg']"]
    veg_indicators = ['Veg', 'Vegetarian', "['Veg']"]
    
    if any(indicator in category for indicator in non_veg_indicators):
        return 'Non-veg'
    elif any(indicator in category for indicator in veg_indicators):
        return 'Veg'
    else:
        return 'Non-veg'  # Handle unexpected or missing values

# Iterate over the dictionary and update the 'v_n_ng' value
for link, info_list in gmaps_to_link.items():
    for info in info_list:
        if not pd.isnull(info["lat"]) and not pd.isnull(info["long"]):
            category = str(info.get('v_n_ng'))
            mapped_category = map_to_veg_nonveg(category)
            info['v_n_ng'] = mapped_category
```

```python
m = folium.Map(location=[12.9, 77.5], zoom_start=7, tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Street_Map/MapServer/tile/{z}/{y}/{x}', attr='Esri')

# Initialize the dictionary of feature groups based on veg/non-veg/other type
category_groups = {}

for link, info_list in gmaps_to_link.items():
    for info in info_list:
        if not pd.isnull(info["lat"]) and not pd.isnull(info["long"]) :
            popup_content = f"<b>{info['rest_name']}</b><br>"
            if 'link' in info:
                popup_content += f"YouTube: <a href='{info['link']}' target='_blank'><i class='fab fa-youtube'></i></a><br>"
            if 'address' in info:
                popup_content += f"Address: {info['address']} <a href='{link}' target='_blank'><i class='fab fa-google'></i></a>"

            # Determine the veg/non-veg/other type and add to the appropriate feature group
            category = str(info.get('v_n_ng'))
            # price = info.get('price', 'Not sure')
            if category not in category_groups:
                category_groups[category] = folium.FeatureGroup(name=category)
                
            
            folium.Marker(
                location=[info["lat"], info["long"]],
                popup=folium.Popup(popup_content, max_width='50%'),
                tooltip=info["rest_name"]
            ).add_to(category_groups[category])

# Add all feature groups to the map
for category, group in category_groups.items():
    group.add_to(m)

# Add a layer control to the map
folium.LayerControl().add_to(m)

LocateControl().add_to(m)

Geocoder().add_to(m)

m
```
